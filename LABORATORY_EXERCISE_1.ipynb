```r
#Exercise 4
orders= c(19, 28, 17, 20, 30, 18, 26, 15, 20, 17, 12, 35, 14)
mean(orders)
median(orders)
quantile(orders)
range(orders)
IQR(orders)
var(orders)
sd(orders)
# Plot
boxplot(orders, main="Boxplot of Orders", ylab="Number of Orders") # [cite: 30]
```
