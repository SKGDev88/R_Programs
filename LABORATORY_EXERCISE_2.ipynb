```r
#Exercise 9
defects=c(4, 0, 3, 3, 2, 3, 5, 8, 6, 6)
#mu = 5 sets the hypothesized mean
test_result = t.test(defects, mu = 5, alternative = "less")
print(test_result)
```

```r
#Exercise 10
#Hypothesis: Is the average daily intake exactly 50 gm?
protein= c(48.1, 52.1, 52.0, 48.4, 49.4, 52.0, 46.6, 49.8, 46.4, 50.2, 48.8, 46.5, 50.0, 51.4, 48.0)
t.test(protein, mu = 50)
```

```r
#Exercise 11
#Hypothesis: Is the batch "superior" (mean > 11.5)?
marks = c(9.5, 10.0, 14.5, 13.5, 14.0, 15.0, 10.5, 12.2, 11.0, 10.5, 14.0, 12.0, 13.5, 12.0, 14.0, 15.0)
t.test(marks, mu = 11.5, alternative = "greater")
```

```r
#Exercise 13
#Hypothesis: Can we assume the mean weight is 65 gm?
leaves= c(63, 63, 64, 65, 66, 69, 69, 70, 70, 71)
t.test(leaves, mu = 65)
```

```r
#Exercise 14
#Hypothesis: Mean score > 70 with 90% confidence ($\alpha = 0.10$).
scores = c(62, 92, 75, 68, 83, 95)
t.test(scores, mu = 70, alternative = "greater", conf.level = 0.90)
```

```r
#Exercise 15
#Hypothesis: Significance of weight increase due to Food B.
foodA=c(49, 53, 51, 52, 47, 50, 52, 53)
foodB=c(52, 55, 52, 53, 50, 54, 54, 53)
t.test(foodA, foodB, alternative = "less") # Testing if B > A
```

```r
#Exercise 16
#Hypothesis: Does the high-protein group differ from the general population mean (7.0)?
protein_levels = c(7.1, 7.70, 8.2, 7.56, 7.05, 7.08, 7.21, 7.25, 7.36, 6.59, 6.85, 7.9, 7.27, 6.56, 7.93, 8.5)
t.test(protein_levels, mu = 7.0)
```

```r
#Exercise 17
#Hypothesis: Significant difference before and after farm yard manure.
before= c(7.7, 8.5, 7.2, 6.3, 8.1, 5.2, 6.5, 9.4, 8.3, 7.5)
after=c(8.1, 8.9, 7.0, 6.1, 8.2, 8.0, 5.8, 8.9, 8.7, 8.0)
t.test(before, after, paired = TRUE)
```
